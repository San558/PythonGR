```python
# Functions

def myfun(arg1,arg2):
    print("This is a function")
    res = arg1 + arg2
    return res

z = myfun(2,3)
print(z)
```

```python
def myfun(arg1 = 7,arg2 = 7):
    print("This is a function")
    res = arg1 + arg2
    return res

z = myfun()
print(z)
z2 = myfun(5)
print(z2)
z1 = myfun(arg2 = 3)
print(z1)
z3 = myfun(5,3)
print(z3)
```
