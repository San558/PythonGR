```python
'''Develop a program to trigger the above progress bar example at specified time stamp 2021-03-13 08:55:00'''

import itertools
import time
import datetime
pbo = itertools.cycle(['-', '//', '|', '/'])
count = 20

def progressBar():  
    count = 20
    while count:
        print(f'0-Progressing...[{next(pbo)}]', end = '\r')
        count = count-1
        time.sleep(0.3)
        
print('Starting soon')
wait = True
while wait:
    d_t = str(datetime.datetime.now()).split('.')[0]
    if d_t == '2021-03-13 09:20:40':
        progressBar()
        wait = False
    time.sleep(0.5)
        

    
```

```python
import matplotlib.pyplot as plt

x = [1,2,3,4]
y = [10,7,23,5]
fig, ax = plt.subplots()
ax.plot(x,y,'r')
#ax.plot(x,y,'ro')
ax.set(xlabel = 'x axis', ylabel = 'y axis', title = 'Simple Graph')
ax.grid()
#fig.savefig("simplePlot.png")
plt.show()
```

```python
# Matplotlib Annimation function

import matplotlib.pyplot as plt
import matplotlib.animation as mpanim
from random import randint

fig,ax = plt.subplots()
x_data = []
y_data = []
line, = ax.plot(x_data,y_data)

def init():
    ax.set_xlim(0,50)
    ax.set_ylim(0,50)
    ax.clear
    x_data.clear()
    y_data.clear()
    line.set_data(x_data, y_data)
    return line,

def data_gen():
    for x in range(0,50):
        yield x, randint(0,50)
        
def animate(data):
    x,y = data
    x_data.append(x)
    y_data.append(y)
    line.set_data(x_data, y_data)
    return line,

ani = mpanim.FuncAnimation(fig, animate, data_gen, init_func = init, interval = 10, blit = True, repeat = False)
plt.show()
```
