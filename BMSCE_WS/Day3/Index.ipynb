```python
def simpleFunction(**kwargs): #No. of args not defined
    print("Code is executing")
    print(len(kwargs))
    print(kwargs['arg_1'])

simpleFunction(arg_1=10, arg_2=20)
```

```python
#Lambda Functions -> Anonymous function/no name, single expression
addLambda = lambda a,b : a+b  #lambda args:expression
print(addLambda(10,30))
```

```python
#Masking -> used in configuration of MC gpio
maskBit = lambda dataByte, bitPos : dataByte & ~(0x01 << bitPos)
print(hex(maskBit(0xFF, 0)))
print(hex(maskBit(0xFF, 7)))
```
