```python
'''Assignment3: Calculation of Inductance of a Colpitt Oscilator taking frequecy
and capacitance value as input'''

import math

print("Calculation of Inductance of a Colpitt Oscilatore")
pi = math.pi
f = int(input("Enter the frequency:"))
C = int(input("Enter the capacitance value:"))
L = 1/(4*((pi*f)**2)*C)
print("The inductance of the oscillator is:",L)
```

```python
'''Assignment4: Consider binary data bit sequence of '10110100', develop a Python code to 
plot the clocking sequence of mentioned data bits out a serial shift register,
use matplot lib.'''


import matplotlib.pyplot as plt

x = [1,0,1,1,0,1,0,0]
div = 100
l = len(x)
x_bits = []
y = []
for i in range (0,len(x)):
    for j in range (0,div):
        x_bits.append(x[i])
        
for i in range (0,len(x)):
    for j in range(1,div+1):
        y.append(i+(j*0.01))
        
        
plt.plot(y, x_bits, label="Bits")

plt.legend()

plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")

plt.title("Sequence Plot")

plt.show()
        

    
      



```
